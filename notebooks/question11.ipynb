```python
def compute_mean(numbers):
    mean_value = sum(numbers) / len(numbers)
    print(mean_value)

compute_mean([2, 4, 6, 8])
```

The code above takes a list of numbers and outputs the mean or _average_ by dividing the sum of the numbers by the total quantity of numbers provided.

The mathematical equation for finding a mean is:
$$
\bar{x} = \frac{x_1 + x_2 + \dots + x_n}{n}
$$

It can also be written using summation notation:
$$
\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i
$$
